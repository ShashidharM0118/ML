```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your CSV file is named 'data.csv' and has columns 'Week' and 'Sales'
data = pd.read_csv('/content/data.csv')
# Calculate the mean of x and y
x_mean = data['x(weeks)'].mean()
y_mean = data['y(sales in thousands)'].mean()

# Calculate the sum of squares of x and y
x_sum_sq = ((data['x(weeks)'] - x_mean)**2).sum()
xy_sum_prod = ((data['x(weeks)'] - x_mean) * (data['y(sales in thousands)'] - y_mean)).sum()
 
slope = xy_sum_prod / x_sum_sq
intercept = y_mean - (slope * x_mean)


print("Intercept:", round(intercept, 2))
print("Slope:", round(slope, 2))
 
week_5_sales = intercept + (slope * 5)
week_7_sales = intercept + (slope * 7)

print("Predicted sales for week 5:", round(week_5_sales, 2))
print("Predicted sales for week 7:", round(week_7_sales, 2))

# Create a scatter plot of the data
plt.scatter(data['x(weeks)'], data['y(sales in thousands)'], label='Actual Sales')

# Create a line representing the regression equation
x_values = [min(data['x(weeks)']), max(data['x(weeks)'])]
y_values = [intercept + (slope * x) for x in x_values]
plt.plot(x_values, y_values, color='red', label='Regression Line')

 
plt.xlabel('Week')
plt.ylabel('Sales (in thousands)')
plt.title('Sales vs. Week Regression')
 
plt.legend()

 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Assuming your CSV file is named 'data.csv' and has columns 'Week' and 'Sales'
data = pd.read_csv('/content/data.csv')

# Prepare the data for linear regression
X = data['x(weeks)'].values
y = data['y(sales in thousands)'].values

# Create the design matrix (add a column of 1s for the intercept)
X_matrix = np.vstack((np.ones(len(X)), X)).T

# Calculate the coefficients using the matrix method
coefficients = np.linalg.inv(X_matrix.T @ X_matrix) @ X_matrix.T @ y

# Extract the intercept and slope
intercept = coefficients[0]
slope = coefficients[1]

print("Intercept:", round(intercept, 2))
print("Slope:", round(slope, 2))

# Predict sales for the 5th and 7th week
week_5_sales = intercept + slope * 5
week_7_sales = intercept + slope * 7

print("Predicted sales for week 5:", round(week_5_sales, 2))
print("Predicted sales for week 7:", round(week_7_sales, 2))

# Create a scatter plot of the data
plt.scatter(data['x(weeks)'], data['y(sales in thousands)'], label='Actual Sales')

# Create a line representing the regression equation
x_values = [min(data['x(weeks)']), max(data['x(weeks)'])]
y_values = [intercept + (slope * x) for x in x_values]
plt.plot(x_values, y_values, color='red', label='Regression Line')

# Add labels and title
plt.xlabel('Week')
plt.ylabel('Sales (in thousands)')
plt.title('Sales vs. Week Regression')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
